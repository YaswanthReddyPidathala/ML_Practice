```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data=[[2600,3000,3200,3600,4000],[550000,565000,610000,680000,725000]]
data=pd.DataFrame(data).T
data
```

```python
plt.scatter(data[0],data[1])
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[[0]],data[1])
```

```python
reg.predict([[3900]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(data[0],data[1],marker="+",color="red")
plt.plot(data[0],reg.predict(data[[0]]))
```
