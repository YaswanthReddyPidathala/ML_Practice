```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('titanic.csv')
```

```python
data.head()
```

```python
data=data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis="columns")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['sex_n']=LabelEncoder().fit_transform(data['Sex'])
```

```python
data.head()
```

```python
reg=DecisionTreeClassifier()
```

```python
reg.fit(data[["Pclass","Age","sex_n"]],data["Survived"])
```

```python
pre=reg.predict([[4,3,33]])
```

```python
from sklearn.metrics import confusion_matrix

mat=confusion_matrix(data["Survived"],reg.predict(data[["Pclass","Age","sex_n"]]))
mat
```
